```python
rating=pd.read_csv('rating.csv')
```

```python
rating
```

```python
rating = rating.loc[:,["userId","movieId","rating"]]
rating.head(10)
```

```python
df = pd.merge(movie,rating)
```

```python
df
```

```python
df.shape
```

```python
df = df.iloc[:100000,:]
```

```python
df
```

```python
pivot_table = df.pivot_table(index = ["userId"],columns = ["title"],values = "rating")
pivot_table.head(10)
```

```python
my_movie=pivot_table['Jumanji (1995)']
corr_with_other_movies=pivot_table.corrwith(my_movie)
corr_with_other_movies = corr_with_other_movies.sort_values(ascending=False)
corr_with_other_movies.head(10)
```

```python
my_movie1=pivot_table['Father of the Bride Part II (1995)']
corr_with_other_movies=pivot_table.corrwith(my_movie1)
corr_with_other_movies = corr_with_other_movies.sort_values(ascending=False)
corr_with_other_movies.head(10)
```
